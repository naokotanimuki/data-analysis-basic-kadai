```python
import matplotlib.pyplot as plt
import numpy as np
import japanize_matplotlib

fig, axes = plt.subplots(2, 3, tight_layout=True)

#
x = [0, 1, 2, 3]
y = [1, 2, -5, 2]
axes[0, 0].plot(x, y)
axes[0, 0].set_title("")

#sin
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)

axes[0, 1].plot(x1, y1, linewidth=2.0)
axes[0, 1].set(
    xlim=(0, 10), xticks=np.arange(0, 10, 1),
    ylim=(-1, 5), yticks=np.arange(-1, 6))
axes[0, 1].set_title("sin")

#
x2 = np.random.normal(15, 5, 2000)
axes[0, 2].hist(x2)
axes[0, 2].set(
    xlim=(0, 40), xticks=np.arange(0, 40, 20),
    ylim=(0, 600), yticks=np.arange(0, 600, 200)
)
axes[0, 2].set_title("")

#
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))

axes[1, 0].scatter(x3, y3)
axes[1, 0].set(
    xlim=(-2, 10), xticks=np.arange(0, 10, 5),
    ylim=(-2.5, 9), yticks=np.arange(0, 9, 2.5)
)
axes[1, 0].set_title("")

plt.show()
```
